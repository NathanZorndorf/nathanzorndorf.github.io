```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
import pandas as pd
df=pd.read_csv('/Users/naekid/Downloads/infectionrisk.txt', delimiter='\t')
df = df[df['Stay']<=14]
df = pd.concat([df, pd.get_dummies(df['Region'])], axis=1);
region_mapping = {1:'NE',2:'NC', 3:'S', 4:'W'}
df.rename(columns=region_mapping).head()
df = df.rename(columns=region_mapping)
```

```python
df.head()
```

```python
model_1 = ols('InfctRsk ~ Stay + Xray + C(Region)', data=df).fit()
```

```python
model_1.summary()
```

```python
model_2 = ols('InfctRsk ~ Stay + Xray', data=df).fit()
sm.stats.anova_lm(model_2, model_1)
```

```python
model_3 = ols('InfctRsk ~ Stay + Xray + W', data=df).fit()
sm.stats.anova_lm(model_3, model_1)
```
