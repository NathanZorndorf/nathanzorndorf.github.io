Because the above plot rounds the points, you have no sense of how many rows (i.e. cars) are represented by each point (each dot can represent many cars, or a single car).

We can add jitter to get a sense of the density. 

````r
ggplot(mpg, aes(cty, hwy, color = class)) +
    geom_jitter()
````

Another alternative, which may show the density more clearly, is to use geom_count.

`geom_count` this is a variant geom_point() that counts the number of observations at each location, then **maps the count to point area.**

````r
ggplot(mpg, aes(cty, hwy)) +
    geom_count() 
````
